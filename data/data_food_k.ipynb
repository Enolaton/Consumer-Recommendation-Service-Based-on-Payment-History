```python
import pandas as pd
import glob
import os

DATA_PATH = "../food/*.csv"
file_list = sorted(glob.glob(DATA_PATH))

print(file_list)
print(len(file_list))

print(os.getcwd())
print(os.listdir(".."))
print(file_list)
print(len(file_list))
```

```python
df_list = []

for file in file_list:
    # 파일명에서 연월 추출
    year_month = os.path.basename(file).replace(".csv", "")
    
    df = pd.read_csv(file)
    df['year_month'] = year_month  # 핵심 컬럼
    
    df_list.append(df)
    
df_all = pd.concat(df_list, ignore_index=True)
```
