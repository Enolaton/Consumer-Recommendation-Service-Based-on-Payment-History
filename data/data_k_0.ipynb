```python
import pandas as pd
import glob
import os

base_path = ".."
folders = glob.glob(os.path.join(base_path, "카드소비 데이터_*"))

dfs = {}  # {YYYYMM: DataFrame}

# 제거할 컬럼들
DROP_COLS = ["cty_rgn_no", "card_tpbuz_cd"]

for folder in folders:
    csv_files = glob.glob(os.path.join(folder, "*.csv"))
    if not csv_files:
        continue
    
    df_list = []
    for f in csv_files:
        # 인코딩 충돌 처리
        try:
            df = pd.read_csv(f, encoding='cp949')
        except UnicodeDecodeError:
            df = pd.read_csv(f, encoding='utf-8')
        
        # 필터링
        if "card_tpbuz_nm_1" in df.columns:
            df = df[df["card_tpbuz_nm_1"] == "음식"]
        
        df_list.append(df)
    
    if not df_list:
        continue
    
    df = pd.concat(df_list, ignore_index=True)
    
    # 폴더명에서 YYYYMM 추출
    yyyymm = os.path.basename(folder).split("_")[-1]
    
    # 컬럼 제거 (존재하는 컬럼만 안전하게)
    df = df.drop(columns=DROP_COLS, errors="ignore")
    
    dfs[yyyymm] = df

print(dfs["202201"].head())
```

```python
output_path = "./output"
os.makedirs(output_path, exist_ok=True)  # output 폴더가 없으면 생성

for yyyymm, df in dfs.items():
    file_name = f"{yyyymm}.csv"
    file_path = os.path.join(output_path, file_name)
    df.to_csv(file_path, index=False, encoding="utf-8-sig")  # 한글 깨짐 방지
```
